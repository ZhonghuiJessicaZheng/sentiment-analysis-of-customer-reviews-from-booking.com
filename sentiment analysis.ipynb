```python
#import nltk package
import nltk
import pandas as pd
from nltk.sentiment import SentimentAnalyzer
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
f = open('fashion.csv','r')
fashion = pd.read_csv(f,sep=',')
f.close()
fashion.head()
```

```python
#extract reviews
review = fashion['review text']
review
```

```python
posList = []
negList = []
neuList = []
    
sid = SentimentIntensityAnalyzer()
for sentence in review:
    ss = sid.polarity_scores(sentence)
    for s in review:
        pos=ss['pos']
        neg=ss['neg']
        neu=ss['neu'] 

        
    posList.append(pos)
    negList.append(neg)
    neuList.append(neu)
```

```python
fashion['posList'] = posList
fashion['negList'] = negList
fashion['neuList'] = neuList
fashion
```

```python
#write result into csv
fashion.to_csv('assi3 sentiment result1.csv')
```
